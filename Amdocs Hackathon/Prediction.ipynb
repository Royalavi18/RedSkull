```python
datasets['TotalCharges']=datasets['TotalCharges'].astype(float)
```

```python
datasets['gender']=datasets['gender'].replace({'Male':1,'Female':0})
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
datasets['Partner']=datasets['Partner'].replace({'Yes':1,'No':0})
```

```python
datasets['Dependents']=label_encoder.fit_transform(datasets['Dependents'])
```

```python
datasets['PhoneService']=datasets['PhoneService'].replace({'Yes':1,'No':0})
```

```python
datasets['PhoneService'].unique()
```

```python
datasets['MultipleLines']=datasets['MultipleLines'].replace({'Yes':1,'No':0})
```

```python
datasets['MultipleLines'].unique()
```

```python
datasets.corr()
```

```python
datasets['InternetService'].unique()
```

```python
datasets['InternetService']=datasets['InternetService'].replace({'DSL' : 1,'Fiber optic' : 2 ,'No' : 0})
```

```python
datasets['OnlineSecurity'].unique()
```

```python
datasets['OnlineSecurity']=datasets['OnlineSecurity'].replace({'Yes':1,'No':0})
```

```python
datasets['OnlineBackup'].unique()
```

```python
datasets['OnlineBackup']=datasets['OnlineBackup'].replace({'Yes':1,'No':0})
```

```python
datasets['DeviceProtection'].unique()
```

```python
datasets['DeviceProtection']=datasets['DeviceProtection'].replace({'Yes':1,'No':0})
```

```python
datasets['TechSupport']=datasets['TechSupport'].replace({'Yes':1,'No':0})
```

```python
datasets['TechSupport'].unique()
```

```python
datasets['PaymentMethod'].unique()
```

```python
columns=['StreamingTV','StreamingMovies','PaperlessBilling']

for i in columns:
    datasets[i]=datasets[i].replace({'Yes':1,'No':0})
```

```python
datasets['Contract'].unique()
```

```python
datasets['Contract']=datasets['Contract'].replace({'Month-to-month' : 0,'One year':1, 'Two year':2})
```

```python
datasets['Contract'].unique()
```

```python
datasets=datasets.drop(['PaymentMethod'],axis='columns')
```

```python
from sklearn.preprocessing import StandardScaler
standard=StandardScaler()
columns_for_fitting=['tenure','MonthlyCharges','TotalCharges']

datasets[columns_for_fitting]=standard.fit_transform(datasets[columns_for_fitting])
```

```python
datasets.corr()
```

```python
y=datasets['Churn']
```

```python
X=datasets.drop(['Churn','customerID','gender','PhoneService','MultipleLines','OnlineBackup','StreamingMovies','StreamingTV'],axis='columns')
```

```python
y.head()
```

```python
from sklearn.model_selection import  train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.metrics  import accuracy_score
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()

log_model.fit(X_train,y_train)
y_pred1=log_model.predict(X_test)

print(round(accuracy_score(y_pred1,y_test)*100,2))
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model=DecisionTreeClassifier()

tree_model.fit(X_train,y_train)
y_pred2=tree_model.predict(X_test)

print(round(accuracy_score(y_pred2,y_test)*100,2))
```

```python
tree_model.feature_importances_
```

```python
X_train.columns
```

```python
from sklearn.svm import SVC
svc_model=SVC(kernel='rbf',random_state=50)

svc_model.fit(X_train,y_train)
y_pred3=svc_model.predict(X_test)

print(round(accuracy_score(y_pred3,y_test)*100,2))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=20,criterion='entropy',random_state=0)

rf_model.fit(X_train,y_train)
y_pred4=rf_model.predict(X_test)

print(round(accuracy_score(y_pred4,y_test)*100,2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier()

knn_model.fit(X_train,y_train)
y_pred5=knn_model.predict(X_test)

print(round(accuracy_score(y_pred5,y_test)*100,2))
```

```python
from sklearn.metrics import confusion_matrix
cm_rf=confusion_matrix(y_pred4,y_test)
cm_rf
```
